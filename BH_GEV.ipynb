```python
%matplotlib notebook
#import generals
import matplotlib
import matplotlib. pyplot as plt 
import numpy as np
import seaborn as sns
import pandas as pd

#import specific packages
import mne
mne.set_config('MNE_BROWSER_BACKEND', 'qt')
from mne.io import read_raw_eeglab
from pycrostates.datasets import lemon
from pycrostates.cluster import ModKMeans

#import bokeh stuff
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

#define which sub_id we are analysing
sub_id = "010017"
```

```python
#define function to calculate GEV
def calculate_GEV(sub_id, condition):
    mne.set_config('MNE_BROWSER_BACKEND', 'qt')

    raw_fname = lemon.data_path(subject_id=sub_id, condition=condition)
    raw = read_raw_eeglab(raw_fname, preload=True)

    raw.pick('eeg')
    raw.set_eeg_reference('average')

    n_list = range(1, 11)

    GEV_values = []  # Create an empty list to store the GEV values

    df = pd.DataFrame(columns=['GEV' + str(sub_id)])  # Create an empty DataFrame with a 'GEV' column

    for n in n_list:
        n_clusters = n
        ModK = ModKMeans(n_clusters=n_clusters)
        ModK.fit(raw.crop(0, 10), n_jobs=6)
        GEV_value = ModK.GEV_  # Get the GEV value for the current iteration
        df = df.append({'GEV' + str(sub_id): GEV_value}, ignore_index=True)  # Append the GEV value to df2

    return df
```

```python
#
result = calculate_GEV(sub_id= sub_id, condition='EC')
```
