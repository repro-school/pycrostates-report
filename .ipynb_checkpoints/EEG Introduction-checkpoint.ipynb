````python
import mne
mne.set_config('MNE_BROWSER_BACKEND', 'qt')

from mne.io import read_raw_eeglab

from pycrostates.datasets import lemon

raw_fname = lemon.data_path(subject_id='010004', condition='EO')
raw = read_raw_eeglab(raw_fname, preload=True)

raw.pick('eeg')
raw.set_eeg_reference('average')
````

A interactive EEG viewer:https://github.com/mne-tools/mne-qt-browser

```
pip install "mne>=1.0" matplotlib mne-qt-browser
pip install PyQt5
```

````python
raw.plot()
````

````python
from pycrostates.cluster import ModKMeans

n_clusters = 5
ModK = ModKMeans(n_clusters=n_clusters)
````

````python
ModK.fit(raw.crop(0,10), n_jobs=6)
````

````python
ModK.plot()
````
